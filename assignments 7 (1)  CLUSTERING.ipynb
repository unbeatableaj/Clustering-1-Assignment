```python
from sklearn.preprocessing import MinMaxScaler
normalize=MinMaxScaler()
airlines1=normalize.fit_transform(airlines.to_numpy())
airlines2=pd.DataFrame(airlines1,columns=airlines.columns)
airlines2
```

```python
plt.figure(figsize=(16,8))
dendrogram=sch.dendrogram(sch.linkage(airlines2,method='average'))
```

```python
dendrogram=sch.dendrogram(sch.linkage(airlines2,method='ward'))
```

```python
hc=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='average')
```

```python
y_hc=hc.fit_predict(airlines)
clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
clusters
```

```python
airlines2["h_clusterid"]=clusters
```

```python
airlines2.sort_values("h_clusterid")
```

# Kmeans

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
airlines3=scaler.fit_transform(airlines2)
```

```python
airlines3
```

```python
kmeans=KMeans(n_clusters=5,random_state=0)
kmeans.fit(airlines3)
```

```python
kmeans.inertia_
```

```python
wcss=[]
for i in range(1,30):
    kmeans=KMeans(n_clusters=i,random_state=0)
    kmeans.fit(airlines3)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,30),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
cluster_airline=KMeans(5,random_state=34)
cluster_airline.fit(airlines3)
```

```python
cluster_airline.labels_
```

```python
cluster_airline.cluster_centers_
```

```python
airlines4=pd.DataFrame(cluster_airline.labels_,columns=['cluster'])
airlines4
```

```python
airlines5=pd.concat([airlines,airlines4],axis=1)
airlines5
```
